### Read first carefully the following discussion that can help your in solving Question 1.

Python has a set of built-in methods and `__call__` is one of them. The `__call__` method enables Python programmers to write classes where the instances behave like functions and can be called like a function. When the instance is called as a function; if this method is defined, `x(arg1, arg2, ...)` is a shorthand for `x.__call__(arg1, arg2, ...)`.In the code below, the `__call__` method computes an approximation of the derivative
of a function:

```python
class Derivative(object):
    def __init__(self, f, h=1E-5):
        # 1E-5 = 1 x 10^-5, this is a way of computing lim as h approaches 0 of (f(x+h) - f(x))/h--the limit defn of the derivative
        self.f = f
        self.h = float(h)

    def __call__(self, x):
        f, h = self.f, self.h       
        return (f(x+h) - f(x))/h
```

```python
# Calculating the derivative of sin(x) at x=pi.
from math import sin, pi
df = Derivative(sin)
x = pi
df(x)
```

Another useful special method is `__str__`. It is called when a class instance needs to be converted to a string. This happens when we say print a, and a is an instance. Python will then look into the a instance for a `__str__` method, which is supposed to return a string. If such a special method is found, the returned string is printed, otherwise just the name of the class is printed. See below for an example how this can be implemented: 

```python
class Y(object):
    def __init__(self, v0):
        self.v0 = v0
        self.g = 9.81

    def __str__(self):
        return f'{self.v0} - 0.5*{self.g}' 
```

```python
y = Y(1.5)
str(y)
```

1) (5 points) Initialize my polynomial using a list of its coefficients: e.g. `p = Polynomial([1.0, 2.0, 3.0])`

```python
#create Polynomial class
class Polynomial(object):
# define __init__ function for the class
    def __init__(self, xs):
        self.a = xs[0]
        self.b = xs[1]
        self.c = xs[2]
# define the __call__ function for the class which will allow the class Polynomial to act as a function   
    def __call__(self, x):
        a, b, c = self.a, self.b, self.c
        return a*(x**0) + b*(x**1) + c*(x**2)
# define the __str__ class which will return a string representation of the polynomial while ensuring that negative values of coefficients appear properly    
    def __str__(self):
        if self.b >= 0 and self.c >= 0:
            return f'{self.a}x^0 + {self.b}x^1 + {self.c}x^2'
        elif self.b >= 0 and self.c < 0:
            return f'{self.a}x^0 + {self.b}x^1 - {-self.c}x^2'
        elif self.b < 0 and self.c >= 0:
            return f'{self.a}x^0 - {-self.b}x^1 + {self.c}x^2'
        else:
            return f'{self.a}x^0 - {-self.b}x^1 - {-self.c}x^2'
# use Polynomial class to define the given polynomial and display it as a string 
p = Polynomial([1.0, 2.0, 3.0])
str(p)
```

4) (20 points) Add polynomials to get a new polynomial. 

```python
class Polynomial(object):
    
    def __init__(self, xs):
        self.a = xs[0]
        self.b = xs[1]
        self.c = xs[2]
    
    def __call__(self, x):
        a, b, c = self.a, self.b, self.c
        return a*(x**0) + b*(x**1) + c*(x**2)
    
    def __str__(self):
        if self.b >= 0 and self.c >= 0:
            return f'{self.a}x^0 + {self.b}x^1 + {self.c}x^2'
        elif self.b >= 0 and self.c < 0:
            return f'{self.a}x^0 + {self.b}x^1 - {-self.c}x^2'
        elif self.b < 0 and self.c >= 0:
            return f'{self.a}x^0 - {-self.b}x^1 + {self.c}x^2'
        else:
            return f'{self.a}x^0 - {-self.b}x^1 - {-self.c}x^2'
# define a function for the polynomials which modifies the + operator (i.e. the function adds together the coefficients of the polynomial to get a new poly)    
    def __add__(self, q):
        return Polynomial([self.a + q.a, self.b + q.b, self.c + q.c])
# define two new polynomials and then add them together using the function created above
l = Polynomial([1.0, 2.0, 3.0]) + Polynomial([2.0, 4.0, 6.0])
str(l)
```

```python
# import sympify from sympy in order to display the polynomial in the desired manner
from sympy import sympify
from sympy import *
init_printing(order = 'rev-lex')
# define p again for clarity
p = Polynomial([1.0, 2.0, 3.0])
# generate a string representation of the polynomial and assign it to the variable poly_repr
poly_repr = str(p)
# update the polynomial representation so that python will recognize that "ax" is "a*x"
updated_poly_repr = poly_repr.replace("x", "*x")
# update polynomial representation again so that x^0 is not displayed in the final representation
final_poly_repr = updated_poly_repr.replace("*x^0", "")
# print the final string polynomial representation
print(final_poly_repr)
# use sympify to display the final polynomial representation as an UNEVALUATED expression
sympyfied_poly = sympify(final_poly_repr, evaluate = False)
sympyfied_poly
```

The TOTAL column (column D) is set to the Excel formula `=ROUND(B3*C3, 2)`, which multiplies the cost per pound by the number of pounds sold and rounds the result to the nearest cent.  With this formula, the cells in the TOTAL column will automatically update themselves if there is a change in column B or C.

Let's assume that the prices of garlic, celery, and lemons were entered incorrectly, which would require someone to go through thousands of rows in the spreadsheet to update the cost per pound for any garlic, celery, and lemon rows.  You can't do a simple find-and-replace for the price because there may be other items with the same price that you don't want to update by mistake.  Your code should do the following:

 - Loop over all the rows.
 - For each row, if the value in column A is garlic, celery, or lemons, update the price in column B.

You may want to store the corrected price information in a dictionary and write your code to use this data structure.

```python
# Write your code below
# import the necessary library
import xlwings as xw
# create a Book object that points to the updatedProduceSales.xlsx file
wb = xw.Book("updatedProduceSalesNew.xlsx")
# grab the data from Sheet1
data = wb.sheets("Sheet1")
# create a list of the names of the produce
produce_names = list(data.range("A2", "A23758").value)
# create a list of the prices of the produce
produce_prices = list(data.range("B2", "B23758").value)
# loop through the two lists in order to replace the desired prices
for i in range(len(produce_names)):
    if produce_names[i] == "Celery":
        produce_prices[i] = 1.19
    if produce_names[i] == "Garlic":
        produce_prices[i] = 3.07
    if produce_names[i] == "Lemon":
        produce_prices[i] = 1.27
# create a produce_list which consists of sublists that include the name and price of each produce type
produce_list = list(zip(produce_names, produce_prices))
# print produce_list to check work
print(produce_list)
# clear the name and price data from the excel sheet 
data.range('A2:B23578').clear_contents()
# insert the new name and price pairs into the excel sheet
data.range('A2', 'B23578').value = [item for item in produce_list]
```
